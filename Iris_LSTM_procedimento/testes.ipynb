```python
import pickle
import os
import pandas as pd
import numpy as np
dir = 'E:/OneDrive/Documentos/GitHub/iris/Iris_LSTM_procedimento/train_data/'
```

```python
x = pd.read_pickle(dir + 'x.pkl')
y = pd.read_pickle(dir + 'y.pkl')
```

```python
x.columns = x.columns.astype(str)
x
```

```python
y.columns = y.columns.astype(str)
y
```

```python
x_melt = x.melt(id_vars=['cnes', 'codigo_procedimento'], var_name='date', value_name='x')
y_melt = y.melt(id_vars=['cnes', 'codigo_procedimento'], var_name='date', value_name='y')

df = pd.merge(x_melt, y_melt, on=['cnes', 'codigo_procedimento', 'date'])
```

```python
df.to_pickle(dir + 'df.pkl')
```
