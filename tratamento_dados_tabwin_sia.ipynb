```python
import pandas as pd
import numpy as np
import os

# Definindo a opção "float_format" para exibir somente 1 dígito após a vírgula
pd.set_option("float_format", "{:.2f}".format)
```

```python
df_proc_sia_tw = pd.read_csv('./data/tabwin/procedimento_sia.csv', sep= '|')
df_proc_sia_tw
```

```python
df_proc_sia_tw.dtypes
```

```python
# cria um df com base em uma pivot table para os dados de quantidade de procedimento por cnes, com colunas com dados do município
df_proc_mun_tw = (
    df_proc_sia_tw
    .pivot_table(index=['CO_MUNICIPIO_IBGE', 'Município', 'UF', 'CO_PROC_REALIZADO', 'NO_PROCEDIMENTO'], columns='DT_CMPT', values='QTD_SIA')
    .reset_index()
    .sort_index()
)
```
